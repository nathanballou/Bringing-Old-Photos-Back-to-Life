# Image Resoration and Enhacement

To use this, simply upload your picture(s) to the **Upload_Pictures_Here** folder. The tool supports *.jpg*, *.jpeg*, *.tiff*, and *.png* files. Others may work, but have not been tested.

Then just run the three code cells.

Once the restoration is done, the final images will be displayed to the right of the origional images.

To download the final output, simply navigate to **Output/final_output** and download the pictures.

```python
%cd /home/jovyan/git-repos/photo-restoration/

import os
import shutil

import IPython.display
import PIL.Image

from display_functions import make_grid
```

```python
basepath = os.getcwd()
upload_path = os.path.join(basepath, "Upload_Pictures_Here")
output_path = os.path.join(basepath, "Output")

if os.path.isdir(output_path):
    shutil.rmtree(output_path)

os.mkdir(output_path)

!python Code/run.py --input_folder Upload_Pictures_Here --output_folder Output --GPU -1 --with_scratch --HR
```

```python
filenames = os.listdir(upload_path)
filenames.sort()

for filename in filenames:
    print(filename)
    image_original = PIL.Image.open(os.path.join(upload_path, filename))
    image_restore = PIL.Image.open(os.path.join(output_path, "final_output", os.path.splitext(filename)[0])+".png")

    display(make_grid(image_original, image_restore))
```
